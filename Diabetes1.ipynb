```python
import pandas as pd
from tensorflow import keras

data=pd.read_csv('diabetes.csv')

# Split the data into train and test data
train_dataset = data.sample(frac=0.8, random_state=0)
test_dataset = data.drop(train_dataset.index)

X_train = train_dataset.copy()
X_test = test_dataset.copy()

y_train = X_train.pop('Outcome')
y_test = X_test.pop('Outcome')


# Create the model
model =keras.Sequential()
model.add(keras.layers.Dense(8, activation='relu'))
model.add(keras.layers.Dense(4, activation='relu'))
model.add(keras.layers.Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=1)

#save the model and the weights
model.save('./diabetes/diabetes_model.h5')
model.save_weights('./diabetes/weights')
print(model.summary())

```

```python
diabetes_model = keras.models.load_model('diabetes/diabetes_model.h5')
#diabetes_model.summary()
Y=diabetes_model.predict(X_test)

loss, acc = model.evaluate(X_test,y_test)

print("Restored model, accuracy: {:5.2f}%".format(100 * acc))
```
